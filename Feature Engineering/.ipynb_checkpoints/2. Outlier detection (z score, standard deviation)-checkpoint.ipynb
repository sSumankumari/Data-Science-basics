# Outlier detection and removal: z score, standard deviation

```python
import pandas as pd
import numpy as np
import matplotlib
from matplotlib import pyplot as plt
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (10,6)
```

```python
df = pd.read_csv('heights_1.csv')
df.sample(5)
```

```python
df.shape
```

```python
plt.hist(df.height, bins=20, rwidth=0.8)
plt.xlabel('Height (inches)')
plt.ylabel('Count')
# Repesentation is like a Bell curve
plt.show()
```

```python
from scipy.stats import norm

plt.hist(df.height, bins=20, rwidth=0.8, density=True)
plt.xlabel('Height (inches)')
plt.ylabel('Count')

rng = np.arange(df.height.min(), df.height.max(), 0.1)
plt.plot(rng, norm.pdf(rng, df.height.mean(), df.height.std()))
```

```python
df.height.min()
```

```python
df.height.max()
```

```python
df.height.describe()
```

```python
df.height.mean()
```

```python
df.height.std()
```

```python
upper_limit = df.height.mean() + 3*df.height.std()
upper_limit
```

```python
lower_limit = df.height.mean() - 3*df.height.std()
lower_limit
```

### Outlier Detection

```python
df[(df.height>upper_limit) | (df.height<lower_limit)]
```

```python
# Removing the Outliers
df_no_outliers1 = df[(df.height < upper_limit) & (df.height > lower_limit)]
df_no_outliers1
```

```python
df.shape[0] - df_no_outliers1.shape[0]
```

<b> z score indicates how many standard deviation away a data point is. </b>

Calculate the Z score: 
    
    Z = (x - U) / s
    
    U --> mean
    s --> standard deviation

```python
df['zscore'] = (df.height - df.height.mean()) / df.height.std()
df.head()
```

```python
# Checking the z-score
(73.847017 - 66.367560) / 3.847528
```

```python
df[df['zscore'] > 2.8]
```

```python
df[df['zscore'] < -2.8]
```

```python
df[(df.zscore > 2.8) | (df.zscore < -2.8)]
```

```python
df_no_outliers2 = df[(df.zscore < 2.8) & (df.zscore > -2.8)]
df_no_outliers2.head()
```

```python
df.shape[0] - df_no_outliers2.shape[0]
```

#### Task:-

(1) Remove outliers using percentile technique first. Use [0.001, 0.999] for lower and upper bound percentiles

(2) After removing outliers in step 1, you get a new dataframe.

(3) On step(2) dataframe, use 4 standard deviation to remove outliers

(4) Plot histogram for new dataframe that is generated after step (3). Also plot bell curve on same histogram

(5) On step(2) dataframe, use zscore of 4 to remove outliers. This is quite similar to step (3) and you will get exact same result

```python
df1 = pd.read_csv('bhp.csv')
df1.head()
```

```python
df1.shape
```

```python
plt.hist(df1.price_per_sqft, bins=20, rwidth=0.8)
plt.xlabel('Price per square ft')
plt.ylabel('Count')
plt.show()
```

```python
plt.hist(df1.price_per_sqft, bins=20, rwidth=0.8)
plt.xlabel('Price per square ft')
plt.ylabel('Count')
plt.yscale('log')
plt.show()
```

### Treating Outliers by using Percentile 

```python
lower_limit, upper_limit = df1.price_per_sqft.quantile([0.001, 0.999])
lower_limit, upper_limit
```

```python
outliers = df1[(df1.price_per_sqft>upper_limit) | (df1.price_per_sqft<lower_limit)]
outliers.sample(10)
```

```python
df2 = df1[(df1.price_per_sqft<upper_limit) & (df1.price_per_sqft>lower_limit)]
df2.shape
```

```python
df1.shape
```

```python
df1.shape[0] - df2.shape[0]
```

### Now Treating Outliers by using 4 Standard Deviation

```python
max_limit = df2.price_per_sqft.mean() + 4*df2.price_per_sqft.std()
min_limit = df2.price_per_sqft.mean() - 4*df2.price_per_sqft.std()
max_limit, min_limit
```

```python
df2[(df2.price_per_sqft>max_limit) | (df2.price_per_sqft<min_limit)].sample(10)
```

```python
df3 = df2[(df2.price_per_sqft>min_limit) & (df2.price_per_sqft<max_limit)]
df3.shape
```

```python
df2.shape[0] - df3.shape[0]
```

```python
# Hist Plot
plt.hist(df3.price_per_sqft, bins=20, rwidth=0.8)
plt.xlabel('Price per square ft')
plt.ylabel('Count')
plt.show()
```

```python
from scipy.stats import norm
import numpy as np

plt.hist(df3.price_per_sqft, bins=20, rwidth=0.8, density=True)
plt.xlabel('Height (inches)')
plt.ylabel('Count')

rng = np.arange(-5000, df3.price_per_sqft.max(), 100)
plt.plot(rng, norm.pdf(rng,df3.price_per_sqft.mean(),df3.price_per_sqft.std()))
```

### Removing Outliers by using z-score

```python
df2['zscore'] = (df2.price_per_sqft-df2.price_per_sqft.mean())/df2.price_per_sqft.std()
df2.head()
```

```python
df4 = df2[(df2.zscore > -4) & (df2.zscore < 4)]
df4.shape
```

```python
df2.shape[0] - df4.shape[0]
```

Thus, the total number of outliers removed by using any of the two mwthods (either Z-score or standard deviation) are equal to 125.
