# Format Strings in plot function

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x=[1, 2, 3, 4, 5, 6, 7]
y=[50, 51, 52, 48, 47, 49, 46]
```

```python
plt.plot(x,y)
```

```python
# using marker to plot
plt.plot(x,y, 'g+')
```

```python
plt.plot(x,y, 'g+--')
```

```python
plt.plot(x,y, 'r*--')
```

```python
# Keyword argument
plt.plot(x,y, color='lightgrey', marker='D', linestyle='', markersize=10)
```

```python
# plt.plot(x,y, color='blue', alpha=1)
plt.plot(x,y, color='blue', alpha=0.3)
```
