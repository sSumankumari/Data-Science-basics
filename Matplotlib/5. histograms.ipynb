# Histograms

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar)
```

```python
plt.hist(blood_sugar, bins=4)
```

```python
plt.hist(blood_sugar, bins=4, rwidth=0.95)
```

```python
plt.hist(blood_sugar, bins=[80, 100, 125, 200], rwidth=0.95, color='g')
```

```python
plt.hist(blood_sugar, bins=[80, 100, 125, 200], rwidth=0.95, color='g', histtype='step')
```

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]
```

```python
plt.hist([blood_sugar_men, blood_sugar_women], rwidth=0.95, color=['green', 'orange'])
```

```python
plt.xlabel('Sugar Range')
plt.ylabel('Total no. of patients')
plt.title('Blood Sugar Analysis')

plt.hist([blood_sugar_men, blood_sugar_women], rwidth=0.95, color=['green', 'orange'], label=['men', 'women'])
plt.legend()
```

```python
plt.ylabel('Sugar Range')
plt.xlabel('Total no. of patients')
plt.title('Blood Sugar Analysis')

plt.hist([blood_sugar_men, blood_sugar_women], rwidth=0.95, color=['green', 'orange'], label=['men', 'women'], orientation='horizontal')
plt.legend()
```
