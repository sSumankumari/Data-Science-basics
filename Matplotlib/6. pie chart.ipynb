# Pie Chart

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals = [1400, 600, 300, 410, 250]
exp_labels = ['Home rent', 'Food', 'Phone/Internet Bill', 'Car', 'Other utilities']
```

```python
plt.pie(exp_vals, labels=exp_labels)
```

```python
plt.axis('equal') # to draw a perfect circle for the pie chart
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%.2f%%', shadow=True, explode=[0,0,0,0,0])
plt.show()
```

```python
plt.axis('equal') # to draw a perfect circle for the pie chart
plt.pie(exp_vals, labels=exp_labels, radius=1.5, autopct='%.2f%%', shadow=True, explode=[0,0.3,0,0.2,0])
plt.show()
```

```python
plt.axis('equal')
plt.pie(exp_vals, labels=exp_labels, radius=1, autopct='%.2f%%', startangle = 45)
plt.show()
```

# How to save Chart to a file using savefig() function?

```python
plt.pie(exp_vals, labels=exp_labels, radius=1, autopct='%.2f%%', startangle = 45, shadow=True, explode=[0,0.1,0,0.2,0])

plt.savefig('piechart.png', bbox_inches='tight', pad_inches=2, transparent=True)
```
