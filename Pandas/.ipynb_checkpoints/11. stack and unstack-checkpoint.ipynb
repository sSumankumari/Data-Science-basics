# Stack and Unstack in Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel('stock.xlsx', header=[0,1])
df
```

```python
df_stacked=df.stack()
df_stacked
```

```python
df.stack(level=0)
```

```python
# Unstacking of a stacked dataframe gives original dataframe
df_stacked.unstack()
```
