# Pandas Group By feature

```python
import pandas as pd
df = pd.read_csv('cities_weather.csv')
df
```

```python
g = df.groupby('City')
g
```

```python
# To access each of the groups
for city, city_df in g:
    print(city)
    print(city_df)
```

```python
# to access a specific dataframe
g.get_group('Paris')
```

```python
# Direct way of accessing groupby df
df.groupby('Event').get_group('Sunny')
```
