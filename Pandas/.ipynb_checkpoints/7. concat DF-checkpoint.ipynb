# Concat DataFrame

```python
import pandas as pd
```

```python
india_wthr = pd.DataFrame({
    'city' : ['mumbai', 'delhi', 'banglore'],
    'temperature' : [32, 45, 30],
    'humidity' : [80, 60, 78]
})
india_wthr
```

```python
us_wthr = pd.DataFrame({
    'city' : ['new york', 'chicago', 'orlanda'],
    'temperature' : [21, 14, 35],
    'humidity' : [68, 65, 75]
})
us_wthr
```

```python
df = pd.concat([india_wthr, us_wthr])
df
```

```python
df = pd.concat([india_wthr, us_wthr], ignore_index = True)
df
```

```python
df = pd.concat([india_wthr, us_wthr], keys=['India', 'US'])
df
```

```python
df.loc['India']
# df.loc['US']
```

```python
temp_df = pd.DataFrame({
    'city' : ['mumbai', 'delhi', 'banglore'],
    'temperature' : [32, 45, 30]
}, index=[0,1,2])
temp_df
```

```python
humidity_df = pd.DataFrame({
    'city' : ['delhi', 'mumbai'],
    'humidity' : [80, 60]
}, index=[1,0])
humidity_df
```

```python
# To make windspeed column appear as a column in temp_df
new_df = pd.concat([temp_df, humidity_df])
new_df
```

```python
# By default axis=0
new_df = pd.concat([temp_df, humidity_df], axis=1)
new_df
```

```python
s = pd.Series(['Humid', 'Dry', 'Rain'], name='event')
s
```

```python
pd.concat([temp_df, s], axis=1)
```
