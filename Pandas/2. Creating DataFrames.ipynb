```python
# From Python Dictionary
weather_data = {
    'day': ['01-01-2022', '01-02-2022', '01-03-2022', '01-04-2022', '01-05-2022', '01-06-2022', '01-07-2022'],
    'temperature': [32, 35, 28, 24, 32, 32, 37],
    'windspeed': [6, 7, 2, 3, 5, 7, 4],
    'event': ['rain', 'sunny', 'snow', 'snow', 'rain', 'sunny', 'rain']
}
df3 = pd.DataFrame(weather_data)
df3
```

```python
# By using Tuples list
weather = [
    ('01-01-2022',32,6,'Rain'),
    ('01-01-2022',35,7,'Sunny'),
    ('01-03-2022',28,2,'Snow'),
    ('01-04-2022',24,3,'Snow'),
    ('01-05-2022',32,5,'Rain'),
    ('01-06-2022',32,7,'Sunny'),
    ('01-07-2022',37,4,'Rain')
]
df4 = pd.DataFrame(weather, columns=['day', 'temperature', 'windspeed', 'event'])
df4
```

```python
# By creating a list of dictionaries
weather_df = [
    {'day':'01-05-2022', 'temperature':32, 'windspeed':5, 'event':'Rain'},
    {'day':'01-06-2022', 'temperature':32, 'windspeed':7, 'event':'Sunny'},
    {'day':'01-07-2022', 'temperature':37, 'windspeed':4, 'event':'Rain'}
]
df5 = pd.DataFrame(weather_df)
df5
```
