# handle Missing Data : replace() function

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('weather_data2.csv')
df
```

```python
# To replace any special value
new_df = df.replace(32, np.NaN)
new_df
```

```python
# To replace two special values
nw_df = df.replace([32, 'sunny'], np.NaN)
nw_df
```

```python
# To replace value based on specific column
nw_df = df.replace({
    'temperature': '32',
    'event': 'cloudy'
}, np.NaN)
nw_df
```

```python
nw_df = df.replace({
    28.0 : 0,
    'cloudy':'cold'
})
nw_df
```

```python
# regex --> regular expression
# new_df = df.replace('[a-z]', '', regex=True)
new_df = df.replace('[A-Z]', '', regex=True)

new_df
```

```python
new_df = df.replace({
    'temperature': '[A-Za-z]',
    'event': '[A-Z]'
}, '', regex=True)

new_df
```
